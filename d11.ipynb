```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([12,15,19,36,37,51,60,65,71,81])
y=np.array([13,23,34,45,48,56,67,72,81,90])
z=np.array([11,19,25,32,45,51,62,73,80,98])
```

## Line Plot

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))

plt.plot(x,y,marker='o',mfc='white', color='black',ls=':', ms=7);
plt.grid()
```

```python
plt.title('Line plot 2')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,15))
plt.yticks(range(1,90,15))
plt.plot(x,z, marker='*',mfc='red', color='green',ls=':', ms=10);
plt.grid()
#markers= <,>,^,H,o,*
```

```python
plt.title('Line plot 3')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,10))
plt.yticks(range(1,90,10))
plt.plot(x,y,marker='o',mfc='green', color='black',ls=':', ms=7, label= 2023);
plt.plot(x,z, marker='*',mfc='red', color='yellow',ls='--', ms=10,label=2024);
plt.grid()
plt.legend();
```

##### Subplot

```python

plt.figure(figsize=(16,9))
plt.subplot(3,3,2)
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,10))
plt.yticks(range(1,90,10))
plt.plot(x,y,marker='o',mfc='white', color='black',ls=':', ms=7);
plt.grid()

plt.subplot(3,3,1)
plt.title('Line plot 2')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,15))
plt.yticks(range(1,90,15))
plt.plot(x,z, marker='*',mfc='red', color='green',ls=':', ms=10);
plt.grid()
```

# Box plot

```python
marks=[67,65,55,59,69,72,51,70,53,98,23]
```

```python
plt.boxplot(marks);
'''
circles are outliers
box in IQR ( inner )
greater than upper baund and lesss than lower baund
'''
```

```python
upper = np.percentile(marks, 75)
lower = np.percentile(marks, 25)
```

```python
print(upper)
print(lower)
```

```python
upper,inter,lower=np.percentile(marks, [75, 50, 25])
```

```python
print(upper,inter,lower)
```

```python
iqr= upper-lower
iqr
```

```python
ub=upper +(1.5*iqr)
lb= lower + (1.5 * iqr)
print(ub,lb)
```
